#### add primary imports

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import preprocessing
%matplotlib inline


import warnings
warnings.filterwarnings("ignore")
```

#### read csv and investigate datas

```python
df = pd.read_csv('teleCust1000t.csv')
print(df['custcat'].value_counts())
print(df.columns)
df.hist(column='income', bins=50)
plt.show()
```

#### Grouped data to X and Y 
- X : Specifications 
- Y : The column to be predicted

```python
X = df[['region', 'tenure','age', 'marital', 'address', 'income', 'ed', 'employ','retire', 'gender', 'reside']] .values  #.astype(float)
y = df['custcat'].values
```

#### Normalize Data


```python
print(X[0:5])
scaler = preprocessing.StandardScaler()
X = scaler.fit(X).transform(X.astype(float))
X[0:5]
```

#### Train Test Split


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

#### Find best K

```python
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier

Ks = 800                                                # Try KNN between 1 to 800 
mean_acc = np.zeros((Ks-1))
std_acc = np.zeros((Ks-1))

for n in range(1,Ks):
    
    #Train Model and Predict  
    neigh = KNeighborsClassifier(n_neighbors = n).fit(X_train,y_train)
    yhat=neigh.predict(X_test)
    mean_acc[n-1] = metrics.accuracy_score(y_test, yhat)

    
    std_acc[n-1]=np.std(yhat==y_test)/np.sqrt(yhat.shape[0])

mean_acc
```

#### Plot the model accuracy for a different number of neighbors and find the best K



```python
plt.plot(range(1,Ks),mean_acc,'g')
plt.fill_between(range(1,Ks),mean_acc - 1 * std_acc,mean_acc + 1 * std_acc, alpha=0.10)
plt.fill_between(range(1,Ks),mean_acc - 3 * std_acc,mean_acc + 3 * std_acc, alpha=0.10,color="green")
plt.legend(('Accuracy ', '+/- 1xstd','+/- 3xstd'))
plt.ylabel('Accuracy ')
plt.xlabel('Number of Neighbors (K)')
plt.tight_layout()
plt.show()
print( "The best accuracy was with", mean_acc.max(), "with k=", mean_acc.argmax()+1) 
```
